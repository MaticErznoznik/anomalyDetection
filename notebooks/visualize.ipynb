```python
import pandas as pd
import plotly.express as px
```

```python
files = [
    "Braila_nightly_LinearFits.csv",
    "braila_pressure_5770_anomalies.csv",
    "braila_pressure_5771_anomalies.csv",
    "braila_pressure_5772_anomalies.csv",
    "braila_pressure_5773_anomalies.csv",
    "braila_flow_211106H360_anomalies.csv",
    "braila_flow_211206H360_anomalies.csv",
    "braila_flow_211306H360_anomalies.csv",
    "braila_flow_318505H498_anomalies.csv",
    "alicante_salinity_EA003_36_conductivity_anomalies.csv"
]

location = "../log/" + files[6]
```

```python
df = pd.read_csv(location)
df["status_code"] = df["status_code"].astype(str)
df["value"] = df["value"].apply(lambda x: eval(x)[0])
fig = px.scatter(df, x="timestamp", y="value", color="status_code", color_discrete_map={
                "-1": "red",
                "0": "yellow",
                "1": "green",
                "2": "blue"}, 
                title = df["algorithm"][0] + ",   " + str(location[7:-4]))

fig.show()
```
