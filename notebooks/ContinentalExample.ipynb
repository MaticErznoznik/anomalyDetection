```python
import pandas as pd
import plotly.express as px
```

```python
file_num = 49

#30, 1, 2, 

df = pd.read_csv("../data/consumer/Continental_data/" + str(file_num) + ".csv")
display(df)

fig = px.scatter(df,x = 'timestamp', y = 'test_value', color = 'status', title='Continental - ' + str(file_num))
fig.show()
```

```python
def SetColor(x):
    if(x == 1):
        return("green")
    elif(x == 0):
        return("yellow")
    else:
        return("red")


df_border = pd.read_csv("../log/Continental" + str(file_num) + "_borderCheck.csv")
display(df_border)
# make status code as discrete value
df_border["status_code"] = df_border["status_code"].astype(str)

# make graph
fig = px.scatter(df_border, y="value", color=df_border['status_code'], hover_data=['status'])
fig.show()
```
