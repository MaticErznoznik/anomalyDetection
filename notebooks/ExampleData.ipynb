```python
df_filtering = pd.read_csv("../log/latency_Filtering.csv")

# make status code as discrete value
df_filtering["status_code"] = df_filtering["status_code"].astype(str)

# make graph
fig = px.scatter(df_filtering, y="value", color="status_code", hover_data=['status'], color_discrete_sequence=["red", "green", "yellow"])
fig.add_hline(y = 30)
fig.add_hline(y = 60)
fig.show()
```
