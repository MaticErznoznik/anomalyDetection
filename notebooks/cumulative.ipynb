```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/consumer/conductivity_data.csv")

display(df)
values = df["Value"].values[::50]
print(len(values))
#values = np.ones(12000)

#for i in range(8000, len(values)):
#    values[i] += (i-8000)*1
    
values = np.append(values, np.ones(5000)*values[-1])
plt.plot(values)

values = np.convolve(values, np.ones(300)/300, mode='valid')
plt.plot(range(300, 300+len(values)),values)
#plt.ylim([0, 15000])


#t = np.linspace(0, 10, 100)
#values = np.sin(t) + np.random.rand(100)*0.1
```

```python
c_sum = 0
c_sums = [0]
decay = 0.005
for i in range(1, len(values)):
    c_sum *=(1-decay)
    c_sum += np.sign(values[i] - values[i-1])*(np.abs((values[i] - values[i-1])/values[i]))
    c_sums.append(c_sum)
    
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('sample')
ax1.set_ylim([0, 12000])
ax1.set_ylabel('values', color=color)
ax1.plot(values, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('c_sums', color=color)  # we already handled the x-label with ax1
ax2.plot(c_sums, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
